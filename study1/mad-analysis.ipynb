```python
from coba.benchmarks import Result

learners1 = "^(CMT_Memorized\(|CMT_Res|vw|corral_en)"
learners2 = "^(CMT_Memorized\(|CMT_Res|vw|corral_off)"
learners3 = "^(CMT_Memorized\(|CMT_Res|vw|corral_off)"

madish_1 = Result.from_file("./outcomes/madish_1.log").filter_lrn(full_name=learners1).filter_fin() #oldest
madish_2 = Result.from_file("./outcomes/madish_2.log").filter_lrn(full_name=learners2).filter_fin() #old
madish_3 = Result.from_file("./outcomes/madish_3.log").filter_lrn(full_name=learners3).filter_fin() #new
```

```python
madish_1.plot_learners(ylim=(0.25,0.45))
madish_2.plot_learners(ylim=(0.25,0.45))
madish_3.plot_learners(ylim=(0.25,0.45))
```
