```python
rewards = []
base_0_rewards = []
base_1_rewards = []
base_2_rewards = []

for row in interactions.query("simulation_id==122 and learner_id==3 and index < 4000").iterrows():
    rewards.append(row[1].reward)
    #base_0_rewards.append(row[1].reward if row[1].predict_0 >= 0.5 else 1-row[1].reward)
    #base_1_rewards.append(row[1].reward if row[1].predict_1 >= 0.5 else 1-row[1].reward)
    #base_2_rewards.append(row[1].reward if row[1].predict_2 >= 0.5 else 1-row[1].reward)
    base_0_rewards.append(row[1].reward*row[1].predict_0/row[1].predict)
    base_1_rewards.append(row[1].reward*row[1].predict_1/row[1].predict)
    base_2_rewards.append(row[1].reward*row[1].predict_2/row[1].predict)

print(sum(rewards)/len(rewards))
print(sum(base_0_rewards)/len(base_0_rewards))
print(sum(base_1_rewards)/len(base_1_rewards))
print(sum(base_2_rewards)/len(base_2_rewards))

#0.7856*0.0877 + 0.7758*0.0658 + 0.8129*0.8465
0.7524*0.0877 + 0.7454*0.0658 + 0.7828*0.8465
    
```
